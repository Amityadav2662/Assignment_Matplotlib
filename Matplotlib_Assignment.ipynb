```python
# Q2. Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# Note: Also add title, xlabel, and ylabel to the plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
x
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500]) 
plt.subplot(2,2,1)
plt.plot(x,y,color='r')
plt.title("First subplot")


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x1,y1,color='b')
plt.title("Second subplot")

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x2,y2,color="#FF8C00")
plt.title("Third subplot")

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x3,y3,color="#FFFF00")
plt.title("Fourth subplot")
```

```python
The subplot() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.
The subplot() is used for putting the more than one plot in the single plot.
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
Ans.
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights
that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A
bar chart describes the comparisons between the discrete categories. 
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,color='r')
plt.xlabel("Companies")
plt.ylabel("Profits")
plt.title("Vertical Representation")
```

```python
plt.barh(company,profit,color="#FF8C00")
plt.ylabel("Companies")
plt.xlabel("Profits")
plt.title("Horizontal Representation")
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
Ans.
It is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to
visualize the shape of our data. Box plots provide a visual summary of the data with which we can quickly 
identify the average value of the data, how dispersed the data is, whether the data is skewed or not (skewness).
```

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```

```python
plt.boxplot(box2)
plt.show()
```
